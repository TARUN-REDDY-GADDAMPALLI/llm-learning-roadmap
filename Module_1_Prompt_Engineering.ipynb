# Module 1 — My Notes

## Key concepts I learned:
- Temperature 0 = precise (use for code, facts)
- Temperature 1+ = creative (use for writing, ideas)
- System prompt = tells AI what role to play
- Few-shot = give examples before asking
- Chain-of-thought = say "think step by step"
- time.sleep(15) = prevents 429 rate limit error

## Project built:

```python
import time
import google.generativeai as genai

model = genai.GenerativeModel("gemini-2.5-flash")

def ask_llm(prompt, system="You are a helpful assistant.", temperature=0.7, max_tokens=200):
    full_prompt = f"{system}\n\n{prompt}"
    response = model.generate_content(
        full_prompt,
        generation_config=genai.GenerationConfig(
            temperature=temperature,
            max_output_tokens=max_tokens,
        )
    )
    time.sleep(15)  # prevents rate limit error
    return response.text.strip()

print("✅ Ready!")
```

```python
prompt = "Give me a name for a coffee shop open 24 hours"

for temp, label in [(0.0, "Focused"), (0.7, "Balanced"), (1.5, "Creative")]:
    print(f"\n🌡 Temperature {temp} — {label}")
    print("-" * 40)
    print(ask_llm(prompt, temperature=temp))
```

```python
text = "The delivery was late but the product quality was outstanding."

zero = f"Classify sentiment as Positive/Negative/Neutral:\n'{text}'"

few = f"""Classify sentiment. Examples:
"I love this!" → Positive
"Terrible service." → Negative
"It arrived Tuesday." → Neutral

Now classify: "{text}" →"""

cot = f"""Classify sentiment of: "{text}"
Think step by step:
Step 1: Identify emotional words.
Step 2: Are they positive or negative?
Step 3: Give final answer."""

for name, p in [("Zero-shot", zero), ("Few-shot", few), ("Chain-of-Thought", cot)]:
    print(f"\n💬 {name}\n{'─'*40}")
    print(ask_llm(p, temperature=0))
```

```python
import time

question = "What is machine learning?"

personas = [
    ("Normal",      "You are a helpful assistant."),
    ("ELI5",        "Explain everything like I'm 5 years old."),
    ("Shakespeare", "You are Shakespeare. Respond in poetic Early Modern English."),
]

for name, system in personas:
    print(f"\n🎭 {name}\n{'─'*40}")
    print(ask_llm(question, system=system, temperature=0.8))
    time.sleep(15)  # ← wait 15 seconds between calls — stays within 5/min limit
```
